```python
#impotig dataset
df_train = pd.read_csv("train.csv", index_col="PassengerId")
df_test = pd.read_csv("test.csv", index_col="PassengerId")
df_gender = pd.read_csv("gender_submission.csv", index_col="PassengerId")
```

```python
#Fill the NaN vallue of "Age" attribute with mean value i.e. 30
df_train = df_train.fillna({"Age" : 24, "Embarked": "S", "Fare": 0})
df_test = df_test.fillna({"Age" : 24, "Embarked": "S", "Fare":0})
```

```python
#The attribute Cabin has lots of NaN values and Cabin does not make any major impact on the target attribute i.e "Survived"
# So we will remove the Cabin attribute from both the data sets
# We can also remove other attributes which does not impact the target attribute i.e. "Name", "Ticket",
df_train = df_train.drop(["Cabin", "Name", "Ticket"], axis=1)
df_test = df_test.drop(["Cabin", "Name", "Ticket"], axis=1)
```

```python
print("Null values in Train dataset:")
print(df_train.isna().sum())
print()
print("###################################################################################")
print()
print("Null values in Test dataset:")
print(df_test.isna().sum())
```

```python
labEnc = LabelEncoder()
df_train["Embarked"] = labEnc.fit_transform(df_train["Embarked"])
df_train["Sex"] = labEnc.fit_transform(df_train["Sex"])
df_test["Embarked"] = labEnc.fit_transform(df_test["Embarked"])
df_test["Sex"] = labEnc.fit_transform(df_test["Sex"])
```

```python
X_train_df = df_train.iloc[:,1:]
y_train_df = df_train.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_df, y_train_df,  test_size=.20, random_state=1)
```

```python
#Standardisation
stanScale = StandardScaler()
scaled_X_train = stanScale.fit_transform(X_train)
scaled_X_test = stanScale.transform(X_test)
```

### Model Buildeing

```python
modelLogReg = LogisticRegression()
modelLogReg.fit(scaled_X_train, y_train)
```

```python
X_train_pred = modelLogReg.predict(scaled_X_train)
```

```python
x_pred = modelLogReg.predict(scaled_X_test)
```

```python
from sklearn.metrics import accuracy_score
print("AS: ", accuracy_score(y_train, X_train_pred))
print("AS: ", accuracy_score(y_test, x_pred))
```

```python
df_test_scaled = stanScale.fit_transform(df_test)
df_test["Predicted"] = modelLogReg.predict(df_test_scaled)
df_test
```
